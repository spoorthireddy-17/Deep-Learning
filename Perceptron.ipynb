# **Perceptron (Binary Classification)**
Example: Study hours -> pass/fail

```python
import numpy as np
```

# Step 1: Prepare training data

```python
#Training Data
X=np.array([1,2,3,4]) # Study Hours
y=np.array([0,0,1,1]) # Fail(0)/Pass(1)

X, y
```

# Step 2: Initialize Parameters

```python
w=0.0
b=0.0
learning_rate=1
epochs=5

print("Initial Weight:",w)
print("Initial Bias:",b)
```

# Define Activation Function (Step Function)

```python
def step_function(z):
    return 1 if z>=0 else 0
```

# Training Loop

```python
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        #1. Compute Weighted sum
        z = w * X[i] + b
        
        #2.Apply Activation function
        y_pred=step_function(z)
        
        #3. Compute Error
        error= y[i] - y_pred

        #4. Update weight
        w=w+learning_rate*error* X[i]

        #5. Update bias
        b=b+learning_rate*error

        print(f"Input :{X[i]},Predicted value:{y_pred}, actual value:{y[i]},weight:{w},bias:{b}" )
```

# Final Model Parameters

```python
print("\n Final Weight:",w)
print("Final Bias:",b)
```

# Testing the Model

```python
test_value=3.1
z=w*test_value+b
prediction=step_function(z)

print(f"For study hours {test_value}, Prediction", prediction)
```

```python
import matplotlib.pyplot as plt
plt.figure()

# Plot training points
for i in range(len(X)):
    plt.scatter(X[i], y[i])

# Decision boundary
x_values = np.linspace(0, 5, 200)
z_values = w * x_values + b
y_values = np.where(z_values >= 0, 1, 0)

plt.plot(x_values, y_values)

plt.xlabel("Study Hours")
plt.ylabel("Pass(1) / Fail(0)")
plt.title("Perceptron Decision Boundary")
plt.ylim(-0.5, 1.5)

plt.show()
```

# **Perceptron AND Gate**

```python
X=np.array([
[0,0],
[0,1],
[1,0],
[1,1]
])
y=np.array([0,0,0,1])

X,y
```

```python
w=np.zeros(2)
b=0
learning_rate=0.1
epochs=10

print("Initial Weights:",w)
print("Initial Bias:",b)
```

```python
def step_function(z):
    return 1 if z>=0 else 0
```

# Train Perceptron

```python
for epoch in range(epochs):
    print(f"\n Epoch {epoch+1}")

    for i in range(len(X)):
        z=np.dot(w,X[i])+b
        y_pred=step_function(z)
        error=y[i]-y_pred
        
        w=w+learning_rate*error*X[i]
        b=b+learning_rate*error
    
        print(f"Input :{X[i]},Predicted value:{y_pred}, actual value:{y[i]},weight:{w},bias:{b}" )
```

```python
print("\n Final Weight:",w)
print("Final Bias:",b)
```

# Test Model

```python
print("\n Testing AND Gate")
for i in range(len(X)):
    z=np.dot(w,X[i])+b
    prediction=step_function(z)
    print("Input:",X[i],'Prediction:', prediction)
```

```python
plt.figure()

# Plot data points
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], marker='o')
    else:
        plt.scatter(X[i][0], X[i][1], marker='x')

# Decision Boundary
x_values = np.linspace(-0.5, 1.5, 100)

if w[1] != 0:
    y_values = -(w[0]*x_values + b) / w[1]
    plt.plot(x_values, y_values)

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)

plt.xlabel("Input X1")
plt.ylabel("Input X2")
plt.title("Perceptron Decision Boundary - AND Gate")

plt.show()
```

# Perceptron OR Gate

```python
import numpy as np

# OR Gate Dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,1,1,1])

# Initialize
w = np.zeros(2)
b = 0.0
learning_rate = 0.1
epochs = 10

def step_function(z):
    return 1 if z >= 0 else 0

# Training
for epoch in range(epochs):
    print(f"\n Epoch {epoch+1}")
    for i in range(len(X)):
        z = np.dot(w, X[i]) + b
        y_pred = step_function(z)
        error = y[i] - y_pred
        
        w = w + learning_rate * error * X[i]
        b = b + learning_rate * error
        print(f"Input :{X[i]},Predicted value:{y_pred}, actual value:{y[i]},weight:{w},bias:{b}" )

print("\nFinal Weights:", w)
print("Final Bias:", b)

# Testing on same inputs
print("\nPredictions:")
for i in range(len(X)):
    z = np.dot(w, X[i]) + b
    print(f"Input {X[i]}  Output {step_function(z)}")
```

```python
plt.figure()

# Plot class 0 and class 1 separately
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], marker='o')
    else:
        plt.scatter(X[i][0], X[i][1], marker='x')

# Decision boundary: w1*x1 + w2*x2 + b = 0
x_values = np.linspace(-0.5, 1.5, 100)

if w[1] != 0:
    y_values = -(w[0]*x_values + b) / w[1]
    plt.plot(x_values, y_values)

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)

plt.xlabel("Input X1")
plt.ylabel("Input X2")
plt.title("Perceptron Decision Boundary - OR Gate")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset from your table
X = np.array([
    [1,1],  # E1
    [1,0],  # E2
    [0,1],  # E3
    [0,0]   # E4
])

y = np.array([1,1,1,0])

# Initialize
w = np.zeros(2)
b = 0.0
learning_rate = 1
epochs = 10

def step_function(z):
    return 1 if z >= 0 else 0

# Training
for epoch in range(epochs):
    for i in range(len(X)):
        z = np.dot(w, X[i]) + b
        y_pred = step_function(z)
        error = y[i] - y_pred
        
        w = w + learning_rate * error * X[i]
        b = b + learning_rate * error

print("Final Weights:", w)
print("Final Bias:", b)

# ================= Plot =================

plt.figure()

# Plot points
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], marker='o')
    else:
        plt.scatter(X[i][0], X[i][1], marker='x')

# Decision Boundary
x_values = np.linspace(-0.5, 1.5, 100)

if w[1] != 0:
    y_values = -(w[0]*x_values + b) / w[1]
    plt.plot(x_values, y_values)

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)

plt.xlabel("Suspicious Keywords (x1)")
plt.ylabel("Unknown Sender (x2)")
plt.title("Perceptron - Spam Detection (OR Logic)")

plt.show()
```

**Part 4  Critical Thinking**

Now suppose we change:

E3  (0,1) output becomes 0

**Will Perceptron Converge?**

Yes 
Because this dataset is still linearly separable.

Now the logic becomes:

Spam only when:

x = 1

This can be separated by a vertical line:

x = 0.5

So perceptron will converge.

**When Will Perceptron NOT Converge?**

If the dataset becomes like XOR:

This is not linearly separable.

No straight line can separate classes.

Perceptron will:

Keep updating weights

Never stabilize

Never converge

**Limitation of Perceptron**

Perceptron can only solve:

Linearly Separable Problems

It cannot solve:

XOR

Non-linear classification
